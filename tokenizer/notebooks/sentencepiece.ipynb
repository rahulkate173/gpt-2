```python
import sentencepiece as spm
```

```python
with open("toy.txt","w") as f:
  f.write("After getting this base vocabulary, we add new tokens until the desired vocabulary size is reached by learning merges, which are rules to merge two elements of the existing vocabulary together into a new one. So, at the beginning these merges will create tokens with two characters, and then, as training progresses, longer subwords.")
f.close()
```

```python
VOCAB_SIZE = 400

# Define the Llama 2 similar parameters
llama2_spm_params = {
    "input": "toy.txt",
    "model_prefix": "llama2_toy_tokenizer",
    "vocab_size": VOCAB_SIZE,
    "model_type": "bpe",                  # Llama uses BPE
    "byte_fallback": True,                # Decomposes unknown characters into bytes
    "unk_id": 0,                          # <unk> token ID
    "bos_id": 1,                          # <s> token ID
    "eos_id": 2,                          # </s> token ID
    "pad_id": -1,                         # Llama 2 base doesn't define a default pad token ID (-1 disables it)
    "character_coverage": 1.0,            # Full character coverage
    "split_by_whitespace": False,         # Prevents pre-tokenizing by whitespace alone
    "remove_extra_whitespaces": False,    # Keeps exact whitespace structure
}
```

```python
spm.SentencePieceTrainer.train(**llama2_spm_params)
```

```python
sp = spm.SentencePieceProcessor()
sp.load("llama2_toy_tokenizer.model")
vocab = [[sp.id_to_piece(idx)] for idx in range(sp.get_piece_size())]
vocab
```

```python
sp.encode("hello 안녕하세요")
```

```python
tokens = sp.encode("hello 안녕하세요")
print([sp.id_to_piece(idx) for idx in tokens ])
## here the cool thing is as this korean texts e/g: "안녕하세요" in not in "toy.txt" for that the {"byte_fallback":True} will go to byte level and encode it
```
