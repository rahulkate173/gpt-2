```python
from transformers import GPT2LMHeadModel
```

```python
model_hf = GPT2LMHeadModel.from_pretrained("gpt2") # gpt2 = 124M , gpt2-xl = 1.5B
sd_hf = model_hf.state_dict()

for k,v in sd_hf.items():
  print(k,v.shape)
```

```python
## chcek the positional embedding
sd_hf["transformer.wpe.weight"].view(-1)[:20] # first 20
```

```python
# plot all positional embedding
import matplotlib.pyplot as plt
plt.imshow(sd_hf["transformer.wpe.weight"],cmap="gray")
## each row here is one position with 768 dim
```

```python
plt.plot(sd_hf["transformer.wpe.weight"][:,150])
plt.plot(sd_hf["transformer.wpe.weight"][:,200])
plt.plot(sd_hf["transformer.wpe.weight"][:,250])
## here this there are the channels
## every channel as 1024 values refering the seqeuence length
## and their behivour is taken how they are working
```

```python
## let's look at attention weight
plt.imshow(sd_hf["transformer.h.1.attn.c_attn.weight"][:300,:300],cmap="gray")
```
